#### IsolationForest

```python
from sklearn.ensemble import IsolationForest
```

```python
#help(IsolationForest)
```

```python
data_encoded.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data_encoded.head()
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=0.025)
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
```

```python
#-1 for outliers and 1 for inliers.
labels=pd.Series(y_pred_outliers)
labels
```

```python
data["labels"]=labels
data.labels.value_counts()
```

```python
data[data.labels==-1]
```

```python
## Let us add a new data point which is outlier
test=np.array([[5],[3.0],[1.3],[0.25],[1],[0],[0]])
print(test.shape)
clf.predict(test.reshape(1,-1))

```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=0.0025)

clf.fit(data_encoded)
# predictions
y_pred_outliers = pd.Series(clf.predict(data_encoded))
y_pred_outliers.value_counts()
data_encoded["labels"]=y_pred_outliers
data_encoded['scores']=clf.decision_function(data_encoded)
```
