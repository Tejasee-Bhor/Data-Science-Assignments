```python
import numpy as np
from scipy.stats import norm

# Step 1: State the Hypotheses
# Null Hypothesis (H0): The mean weekly operating cost is equal to the theoretical cost.
# Alternative Hypothesis (H1): The mean weekly operating cost is greater than the theoretical cost.

# Given data
sample_mean = 3050  # Sample mean weekly cost in Rs.
theoretical_mean = 4000  # Theoretical mean weekly cost in Rs.
std_dev = 5 * 25  # Standard deviation of weekly cost
sample_size = 25  # Sample size

# Step 2: Calculate the Test Statistic
test_statistic = (sample_mean - theoretical_mean) / (std_dev / np.sqrt(sample_size))

# Step 3: Determine the Critical Value
alpha = 0.05
critical_value = norm.ppf(1 - alpha)  # For a one-tailed test

# Step 4: Make a Decision
if test_statistic > critical_value:
    decision = "Reject the null hypothesis."
else:
    decision = "Fail to reject the null hypothesis."

# Step 5: Conclusion
conclusion = ("Based on the decision in step 4, we conclude that there is not strong evidence "
              "to support the restaurant owners' claim that the weekly operating costs are higher "
              "than the model suggests.")

# Output all the steps
print(f"Step 1: State the Hypotheses\n"
      f"Null Hypothesis (H0): The mean weekly operating cost is equal to the theoretical cost.\n"
      f"Alternative Hypothesis (H1): The mean weekly operating cost is greater than the theoretical cost.\n")

print(f"Step 2: Calculate the Test Statistic\n"
      f"Test Statistic (t) = {test_statistic:.2f}\n")

print(f"Step 3: Determine the Critical Value\n"
      f"Critical Value (Z) = {critical_value:.2f}\n")

print(f"Step 4: Make a Decision\n"
      f"Decision: {decision}\n")

print(f"Step 5: Conclusion\n"
      f"{conclusion}")
```
