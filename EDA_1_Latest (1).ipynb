#### Normalization of the data

```python
def get_stand(x):
    
    return ((x-x.mean())/x.std())
```

```python
x=pd.Series([15,10,12,13,44,5,3,2])

get_stand(x)
```

```python
stand_data=get_stand(data_cleaned4)
stand_data.describe()
```

```python
def get_norm(x):
    return (x-x.min())/(x.max()-x.min())
```

```python
x=pd.Series([10500,10000,12000,13500,40000,50000,30000,20009])

get_norm(x)
```

```python
norm_data=get_norm(data_cleaned4)
norm_data.describe()
```
