```python
#Data Preprocessing# Separate features (X) and labels (y)
X = data.iloc[:, 1:].values  # Assuming the first column is the label
y = data.iloc[:, 0].values
```

```python
# Encode categorical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y_encoded, test_size=0.2, random_state=42)
```

```python
# Step 3: Build and Train the ANN Model
def create_model(hidden_layers=1, neurons_per_layer=128, activation='relu', learning_rate=0.001):
    model = Sequential()
    # Input layer
    model.add(Dense(neurons_per_layer, activation=activation, input_shape=(X_train.shape[1],)))

    # Hidden layers
    for _ in range(hidden_layers - 1):
        model.add(Dense(neurons_per_layer, activation=activation))
        model.add(Dropout(0.2))  # Optional dropout for regularization

    # Output layer (number of classes)
    model.add(Dense(len(np.unique(y_train)), activation='softmax'))

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model

# Basic ANN model training
basic_model = create_model()
basic_model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1, validation_split=0.1)
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import ParameterGrid

# Define hyperparameters grid
param_grid = {
    'hidden_layers': [1, 2, 3],
    'neurons_per_layer': [64, 128, 256],
    'activation': ['relu', 'tanh'],
    'learning_rate': [0.001, 0.01]
}

best_model = None
best_score = 0

```

```python
# Perform grid search
for params in ParameterGrid(param_grid):
    print(f"\nTesting parameters: {params}")
    model = create_model(**params)
    model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=0, validation_split=0.1)
    val_loss, val_accuracy = model.evaluate(X_test, y_test, verbose=0)

    print(f"Validation Accuracy: {val_accuracy}")
    if val_accuracy > best_score:
        best_score = val_accuracy
        best_model = model
```

```python
# Evaluation
# Evaluate the basic model
print("\nBasic Model Evaluation:")
y_pred_basic = np.argmax(basic_model.predict(X_test), axis=1)
print(classification_report(y_test, y_pred_basic))
print("Accuracy:", accuracy_score(y_test, y_pred_basic))

```

```python
# Evaluate the best tuned model
print("\nTuned Model Evaluation:")
y_pred_tuned = np.argmax(best_model.predict(X_test), axis=1)
print(classification_report(y_test, y_pred_tuned))
print("Accuracy:", accuracy_score(y_test, y_pred_tuned))
```

```python
# Discuss performance differences
print("\nPerformance Difference:")
print(f"Accuracy of Basic Model: {accuracy_score(y_test, y_pred_basic)}")
print(f"Accuracy of Tuned Model: {accuracy_score(y_test, y_pred_tuned)}")
```
