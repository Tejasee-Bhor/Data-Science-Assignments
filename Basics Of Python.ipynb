```python
# Exercise 1: Prime Numbers
def is_prime(number):
    """
    Check if a given number is a prime number.
    A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
    """
    if number <= 1:
        return False
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
         return False
    return True
print(is_prime(2)) 
print(is_prime(18)) # Output: False
```

```python
# Exercise 2: Product of Random Numbers
import random

def product_check():
    """
    Generate two random numbers, ask the user for their product, and check if the user's answer is correct.
    """
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    correct_product = num1 * num2
    print(f"Generated numbers are {num1} and {num2}")
    user_product = int(input(f"Enter the product of {num1} and {num2}: "))
    if user_product == correct_product:
        print("Correct! Well done.")
    else:
        print(f"Incorrect. The correct product is {correct_product}.")

# Uncomment to run:
# product_check()
```

```python
# Exercise 3: Squares of Even/Odd Numbers
def print_squares(choice='even'):
    """
    Print the squares of all even or odd numbers within the range of 100 to 200.
    The choice parameter decides whether to print squares of 'even' or 'odd' numbers.
    """
    for num in range(100, 201):
        if choice == 'even' and num % 2 == 0:
            print(f"{num}^2 = {num ** 2}")
        elif choice == 'odd' and num % 2 != 0:
            print(f"{num}^2 = {num ** 2}")

# Choose 'even' or 'odd' based on requirement:
print_squares('even')
print_squares('odd')
```

```python
# Exercise 5: Check for Palindrome
import string

def is_palindrome(s):
    """
    Check if a given string is a palindrome, ignoring spaces, punctuation, and capitalization.
    """
    # Clean the string
    cleaned_str = ''.join(c.lower() for c in s if c not in string.punctuation and not c.isspace())
    return cleaned_str == cleaned_str[::-1]

palindrome_test = "A man a plan a canal Panama"
print(f"'{palindrome_test}' is a palindrome: {is_palindrome(palindrome_test)}")
```
