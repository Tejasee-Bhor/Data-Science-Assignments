```python
# from pandas import read_csv
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
# load data
#filename = 'https://raw.githubusercontent.com/slmsshk/DataSet/main/TSNE_data.csv'
dataframe = pd.read_csv("TSNE_data.csv")
```

```python
dataframe
```

```python
# Split-out validation dataset
array = dataframe.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
dataframe
```

```python
help(TSNE)
```

```python
#TSNE visualization
# from bioinfokit.visuz import cluster
data_tsne = TSNE(n_components=2).fit_transform(X)
cluster.tsneplot(score=data_tsne)
#plt.show()
```

```python
pd.DataFrame(data_tsne)
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right')

#Plot will be stored in the default directory
```
