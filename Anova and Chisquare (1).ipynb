```python
import scipy.stats as stats

p=stats.f_oneway(atkins,gm,sb)[1]
```

```python
h0="means of all the groups are equal"
ha="mean of atleast one group is significantly different"

if p>0.05:
    print(h0)
else:
    print(ha)
```

```python
from scipy.stats.contingency import chi2_contingency
```

```python
import pandas as pd
data=pd.read_csv("chisquare.csv")
data.head()
```

```python
data=pd.crosstab(data.B,data.A).values
data
```

```python
from scipy.stats.contingency import chi2_contingency
chi2_contingency(data)
```

```python
p=chi2_contingency(data)[1]
```

```python
alpha=0.05

print(f"P value is {round(p,4)} ") 
if p<alpha:
    print("smoking and being professional athelete are related to each other (Reject Ho)")
else:
    print("smoking and being professional are independent of each other (Ho holds true)")
```
