```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#un = pd.read_csv("/content/Universities.xlsx")
Univ = pd.read_csv("PCA.csv")
#un.head()
Univ.head()
```

```python
# Considering only numerical data
uni = Univ.iloc[:,1:]
uni.head()
```

```python
# Normalizing the numerical data
std = StandardScaler()
uni_normal=std.fit_transform(uni)
uni_normal
```

```python
pca=PCA()
pca_components=pca.fit_transform(uni_normal)
pca_components=pd.DataFrame(pca_components,columns=["PC1","PC2","PC3","PC4","PC5","PC6"])
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained  Scree plot
plt.plot(var1,color="red")
plt.title('Variance Plot')
plt.show()
```
